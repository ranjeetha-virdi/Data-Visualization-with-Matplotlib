```python
import datetime as datetime

fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(1,1,1)
data=pd.read_csv("SPX.csv",index_col=0,parse_dates=True)
spx=data["SPX"]
spx.plot(ax=ax,style="k-")

#Annotations for both text and arrows arranged appropriatly

crisis_data=[
    (datetime.datetime(2007,10,11),"Peak of bull market"),
    (datetime.datetime(2008,3,12),"Bear Stearns Fails"),
    (datetime.datetime(2008,9,15),"Lehman Bankruptcy")
]


for date,label in crisis_data:
    ax.annotate(label, xy=(date,spx.asof(date)+75),
               xytext=(date,spx.asof(date)+225),
               arrowprops=dict(facecolor="red",headlength=4,headwidth=5,width=2),
               horizontalalignment="left",verticalalignment="top")
    
    
#Zoom in on 2007-20010

ax.set_xlim(["1/1/2007","1/1/2011"])
ax.set_ylim([600,1800])


ax.set_title("Important dates in 2008-2009 financial crisis")
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)


rect=plt.Rectangle((0.2,0.75), 0.4, 0.15, color="k", alpha=0.3)
circ=plt.Circle((0.7,0.2),0.2,color="b",alpha=0.4)
pgon=plt.Polygon([[0.15,0.15],[0.5,0.4],[0.2,0.6]],color="g",alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)


plt.savefig("figpath.png",dpi=400,bbox_inches="tight")
```
